```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class UnDGraph:
    def __init__(self):
        self.edges = []
        self.graph = nx.Graph()

    def addEdge(self, src, dest):
        self.edges.append((src, dest))
        self.graph.add_edge(src, dest)

    def visualize(self):
        plt.figure(figsize=(10, 8))
        pos = nx.spring_layout(self.graph, seed=42)
        nx.draw(
            self.graph,
            pos,
            with_labels=True,
            node_color="lightgreen",
            node_size=2000,
            edge_color="gray",
            font_size=10
        )
        plt.title("Logistics Network Graph")
        plt.show()
    def dfs_recursive(self, node, visited=None):
        if visited is None:
            visited = set()

        print(node, end=" ")
        visited.add(node)

        for neighbor in self.graph.neighbors(node):
            if neighbor not in visited:
                self.dfs_recursive(neighbor, visited)


    def dfs_iterative(self, start):
        visited = set()
        stack = [start]

        while stack:
            node = stack.pop()
            if node not in visited:
                print(node, end=" ")
                visited.add(node)

                for neighbor in self.graph.neighbors(node):
                    if neighbor not in visited:
                        stack.append(neighbor)


```

```python
df = pd.read_csv("logistics_routes.csv")
graph = UnDGraph()

for _, row in df.iterrows():
    graph.addEdge(row[0], row[1])

print("Recursive DFS:")
graph.dfs_recursive("WH1")

print("\n\nIterative DFS:")
graph.dfs_iterative("WH1")

graph.visualize()
```
