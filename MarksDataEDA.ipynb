```python
import pandas as pd
marks = pd.DataFrame({
    "Math": [78,85,90,88,76],
    "Physics": [82,79,88,91,85],
    "Chemistry": [75,80,85,89,90]
})

marks.describe()
```

```python
import matplotlib.pyplot as plt
sales = [120,130,125,140,160,170,180,200,220,250]
plt.hist(sales, color='pink')
plt.show()
```

```python
import seaborn as sns

data = pd.DataFrame({
    "Age": [25,30,35,40],
    "Income": [30000,40000,50000,60000],
    "Spending": [200,300,400,500],
    "Savings": [5000,8000,12000,15000]
})

sns.heatmap(data.corr(), annot=True, color='pink')
plt.show()
```

```python
data["Income"].fillna(data["Income"].mean(), inplace=True)
```

```python
employees = pd.DataFrame({
    "Age": [25,30,35,40,45],
    "Salary": [30000,40000,50000,80000,120000],
    "Experience": [1,3,5,10,20]
})

employees.describe()
employees.boxplot(column="Salary", color='red')
plt.show()
```
