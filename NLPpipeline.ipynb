```python
step1 : clean the text(remove special chars, text converted to single case (generally lower case as number of computations will be lesser), etc)
step 2: tokenization(breaking down sentences into words or tokens based on delimiters)- for tokenization we use regular expressions regex
step 3: stop word removal: words that do not contribute to the meaning of the sentence( it, a etc)
step 4: spelling correction: need correct spellings (aka sun vs son)
step 5: stemming - same and faster than lemmatization, converts word to og form but spelling may be wrong
step 6: lemmatization - words are converted to their original basic present tense form,(aka running-> run, slept->sleep)
step 7: N gram creation(number of times common words are repeated, combination of words, N=3(3 times repeated etc) to find significance of sentence)
step 8: POS tagging(part of sentence tagging aka associate parts of the sentence like each word is a noun, adjective , pronoun etc)
```
