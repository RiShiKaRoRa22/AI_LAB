```python
import networkx as nx
class GraphVisual:
  def __init__(self):
    self.edge=[]
  def addEdge(self, src, dest,weight):
    temp=[src,dest,weight]
    self.edge.append(temp)
  def visualize(self):
    grp= nx.Graph()
    grp.add_weighted_edges_from(self.edge)
    pos= nx.spring_layout(grp)
    nx.draw_networkx_nodes(grp,pos)
    nx.draw_networkx_labels(grp,pos)
    nx.draw_networkx_edges(grp,pos)
    edge_labels=nx.get_edge_attributes(grp,'weight')
    nx.draw_networkx_edge_labels(grp,pos,edge_labels=edge_labels)
    #nx.draw_networkx(grp)
```

```python
G= GraphVisual()
G.addEdge('A','B',weight=1)
G.addEdge('B','C',weight=2)
G.addEdge('A','C',weight=8)
G.visualize()
```
