```python
import numpy as np
import math
import random

ROWS = 6
COLS = 7

PLAYER = 1
AI = 2
EMPTY = 0

DEPTH = 4
```

```python
def create_board():
    return np.zeros((ROWS, COLS))

def print_board(board):
    print(np.flip(board, 0))

def is_valid(board, col):
    return board[ROWS - 1][col] == 0

def get_row(board, col):
    for r in range(ROWS):
        if board[r][col] == 0:
            return r

def drop(board, row, col, piece):
    board[row][col] = piece
```

```python
def win(board, piece):
    for r in range(ROWS):
        for c in range(COLS - 3):
            if board[r][c:c+4].tolist().count(piece) == 4:
                return True

    for c in range(COLS):
        for r in range(ROWS - 3):
            if list(board[r:r+4, c]).count(piece) == 4:
                return True
    # Diag right to left
    for r in range(ROWS - 3):
        for c in range(COLS - 3):
            if [board[r+i][c+i] for i in range(4)].count(piece) == 4:
                return True
    # diag left to right
    for r in range(3, ROWS):
        for c in range(COLS - 3):
            if [board[r-i][c+i] for i in range(4)].count(piece) == 4:
                return True

    return False
```

```python
def evaluate(window, piece):
    score = 0
    opponent = PLAYER if piece == AI else AI

    if window.count(piece) == 4:
        score += 100
    elif window.count(piece) == 3 and window.count(EMPTY) == 1:
        score += 5
    elif window.count(piece) == 2 and window.count(EMPTY) == 2:
        score += 2

    if window.count(opponent) == 3 and window.count(EMPTY) == 1:
        score -= 4

    return score
```

```python
def score_board(board, piece):
    score = 0

    # Center preference
    center = list(board[:, COLS // 2])
    score += center.count(piece) * 3

    # Horizontal
    for r in range(ROWS):
        row = list(board[r, :])
        for c in range(COLS - 3):
            score += evaluate(row[c:c+4], piece)

    # Vertical
    for c in range(COLS):
        col = list(board[:, c])
        for r in range(ROWS - 3):
            score += evaluate(col[r:r+4], piece)

    return score
```

```python
def valid_moves(board):
    return [c for c in range(COLS) if is_valid(board, c)]
```

```python
def minimax(board, depth, maximizing):

    moves = valid_moves(board)
    terminal = win(board, PLAYER) or win(board, AI) or len(moves) == 0

    if depth == 0 or terminal:
        if win(board, AI):
            return None, 100000
        elif win(board, PLAYER):
            return None, -100000
        else:
            return None, score_board(board, AI)

    if maximizing:  # AI turn ai maximizes the min score by player
        best_score = -math.inf
        best_col = random.choice(moves)

        for col in moves:
            row = get_row(board, col)
            temp = board.copy()
            drop(temp, row, col, AI)

            _, score = minimax(temp, depth - 1, False)

            if score > best_score:
                best_score = score
                best_col = col

        return best_col, best_score

    else:  # Player turn- minimising the score

        best_score = math.inf
        best_col = random.choice(moves)

        for col in moves:
            row = get_row(board, col)
            temp = board.copy()
            drop(temp, row, col, PLAYER)

            _, score = minimax(temp, depth - 1, True)

            if score < best_score:
                best_score = score
                best_col = col

        return best_col, best_score
```

```python
board = create_board()
print_board(board)
turn = 0
game_over = False
while not game_over:
    if turn == 0:
        col = int(input("Your move (0-6): "))
        if is_valid(board, col):
            row = get_row(board, col)
            drop(board, row, col, PLAYER)
            if win(board, PLAYER):
                print("You Win!")
                game_over = True
    else:
        col, _ = minimax(board, DEPTH, True)
        row = get_row(board, col)
        drop(board, row, col, AI)
        print("AI plays:", col)

        if win(board, AI):
            print("AI Wins!")
            game_over = True

    print_board(board)
    turn = (turn + 1) % 2
```
